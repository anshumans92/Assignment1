```python
import requests
from datetime import datetime

start_time = 1420070400 # 01-01-2015 at 00:00:00
end_time   = 1420156800 # 01-02-2015 at 00:00:00

response = requests.get("https://api.stackexchange.com/2.2/questions?pagesize=100" +
                        "&fromdate=" + str(start_time) + "&todate=" + str(end_time) +
                        "&order=asc&sort=creation&site=stackoverflow")

print(response) # Displays the HTTP response code (should be 200 for success).
```

```python
import time
def print_creation_dates_json(response):
    """
    Prints the creation_date of all the questions in the response.
    
    Parameters:
        response: Response object
    """
    #to convert the creation date into readable format
    date_format = '%m-%d-%Y %H:%M:%S'
    for item in json_response["items"]:
        print((time.strftime(date_format,time.gmtime(item["creation_date"]))))
```

```python
#start
page_count = 1
#using has_more as per the hint
while json_response["has_more"]:
    time.sleep(1)
    #using the same api call defined in the question code, with the addition of the starting page count
    response = requests.get("https://api.stackexchange.com/2.2/questions?page="+str(page_count)+"&pagesize=100" +
                        "&fromdate=" + str(start_time) + "&todate=" + str(end_time) +
                        "&order=asc&sort=creation&site=stackoverflow")
    json_response = response.json()
    print_creation_dates_json(json_response)
    page_count +=1
```

```python
#I tried using iterparse as per the documentation and the code ran for a really long time. Read up documentation on using minidom for parsing xml to python.
#For high performance xml parsing read https://www.ibm.com/developerworks/xml/library/x-hiperfparse/. This runs for 25 minutes and gives the output for the entire dataset.
import xml.etree.cElementTree as ET
from pandas import *
from lxml import objectify
import pandas as pd
file = open("stackoverflow-posts-2015.xml","r", encoding='UTF-8')
#using etree for xml. From documentation in https://docs.python.org/2/library/xml.etree.elementtree.html
tree = ET.parse(file)
root = tree.getroot()
listing = []

for child in root.findall("row"):
    #using dictionary and .get to get the values as per the hint in the problem
    #Only getting the 4 attributes ID, creationdate, owneruserid and tags
    dictionary = {}
    dictionary['Id'] = child.get('Id')
    dictionary['CreationDate'] = child.get('CreationDate')
    dictionary['OwnerUserId'] = child.get('OwnerUserId')
    Tags = str(child.get('Tags'))
    #Since we want only the first tag, we get T[0]
    if Tags != "None":
        T = Tags.split('>')
        dictionary['Tags'] = T[0]+'>'
    listing.append(dictionary)
    
df = pd.DataFrame(data = listing, columns = ['Id', 'CreationDate', 'OwnerUserId', 'Tags'])
print(df)
```

```python
#We read 4 columns in the previous question and for this one three more are required. 
#I tried appending the already created data frame,but the output was not right, so I created another data frame
import xml.etree.cElementTree as ET
import pandas as pd
from pandas import *
from lxml import objectify
#as per the etree parsing definition
file = open("stackoverflow-posts-2015.xml","r", encoding='UTF-8')
tree = ET.parse(file)
root = tree.getroot()
listing = []
for child in root.findall("row"):
    dictionary = {}
    Post_Type = ""
    PostTypeId = child.get('PostTypeId')
    if PostTypeId == '2':
        dictionary['Post_Type'] = "Answer"
    elif PostTypeId == '1':
        dictionary['Post_Type'] = "Question"
    dictionary['Id'] = child.get('Id')
    dictionary['CreationDate'] = child.get('CreationDate')
    dictionary['OwnerUserId'] = child.get('OwnerUserId')
    Tags = str(child.get('Tags'))
    #to get first tags
    if Tags != "None":
        T = Tags.split('>')
        dictionary['Tags'] = T[0]+'>'
    dictionary['ParentId'] = child.get('ParentId')
    dictionary['Score'] = child.get('Score')
    listing.append(dictionary)
#creating a second dataframe
df2 = pd.DataFrame(data = listing, columns =['Id', 'CreationDate', 'OwnerUserId', 'Tags','ParentId','Score','Post_Type'])
#group by post type
grouped = df2.groupby('Post_Type')
for Post_Type, data in grouped:
    grouped = df2.groupby('Post_Type')
    if Post_Type == 'Answer':
        Answer_data = data[['OwnerUserId', 'Score', 'ParentId']]
    elif Post_Type == 'Question':
        Question_data = data[['Id', 'Tags']]
        
#Using group by as per the hint
grouped_answer = Answer_data.groupby('OwnerUserId')

answer_p = grouped_answer['ParentId'].apply(list)
Total = grouped_answer['Score'].sum()
#to combine the values
total_answer = pd.concat([Total, answer_p], axis=1)
#sort top 100 users by Score
top_100 = total_answer.sort_values(by=['Score'], ascending = False)[:100]

top_100
```

```python
#to get the tags based on the earlier data
empty = []
top_values = top_100.index.values
#data frame based on only answer post types
quest = df2[df2.Post_Type == 'Answer']
new_df = df2[['Id', 'Tags']].set_index('Id')
new_df = new_df.to_dict('dict')
#to iterate over all values
for index in range(len(top_values)): 
    user = top_values[index] 
    #.loc for accessing the rows. Implementation based on information from stack overflow
    df_loc = (quest.loc[quest['OwnerUserId'] == user])['ParentId'] 
    user_list_df = df_loc.tolist()
    tags = []
    empty_string = ""
    for index2 in user_list_df:
        try:
            tag = new_df['Tags'][index2] 
            tags.append(str(tag))
        except:
            pass   
    empty_string = " ".join(tags)
    empty.append(empty_string)  
print(empty)
```

```python
#Using the count vectorizer for the data from http://scikit-learn.org/stable/modules/feature_extraction.html
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(empty)
Y = vectorizer.get_feature_names()
print((Y))
array = X.toarray()
```

```python
#using cosine similarity based on the lecture
from sklearn.metrics.pairwise import cosine_similarity
#gets cosine similarity of each user compared to all 100 users
#keeping heatmap notations
data = []
#same as above definitiona
for index in range(len(array)):
    similar = []
    for index2 in range(len(array)):
        #implementation method noted from stack overflow
        convert = array[index].tolist()
        convert1 = array[index2].tolist()
        temporary = ((cosine_similarity([convert], [convert1])).tolist())
        similar.append(temporary[0][0])
    data.append(similar)
print(data)
```
